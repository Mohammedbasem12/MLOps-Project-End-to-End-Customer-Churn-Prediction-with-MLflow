```python

import pandas as pd
from sklearn.preprocessing import LabelEncoder
# Load the dataset
file_path = '/Users/mohammedbasem/Desktop/WA_Fn-UseC_-Telco-Customer-Churn.csv'
df = pd.read_csv(file_path)
```

```python
df.shape
```

```python
# Display basic info and the first few rows to understand the structure
df.info()
df.head()
```

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
df = df.dropna()
df = df.drop(columns=["customerID"])  # Drop non-informative ID

# Step 2: Identify categorical columns (type = object only)
cat_cols = df.select_dtypes(include="object").columns

# Step 3: Label Encode binary categorical columns
label_encoder = LabelEncoder()
for col in cat_cols:
    if df[col].nunique() == 2:
        df[col] = label_encoder.fit_transform(df[col])

# Step 4: One-Hot Encode categorical columns with more than 2 values
multi_cat_cols = [col for col in cat_cols if df[col].nunique() > 2]
df = pd.get_dummies(df, columns=multi_cat_cols, drop_first=True)
# Step 5: Ensure all columns are numeric (convert bool to int)
df = df.astype(int, errors='ignore')
# Step 5: Save or show
print("Final shape:", df.shape)
df.head()
```
